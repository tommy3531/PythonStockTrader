```python
import json
import requests
import prettytable
import tweepy
import numpy as np
import os
import pickle
```

```python
headers = {'Content-type': 'application/json'}
data = json.dumps({"seriesid": ['CUUR0000SA0','SUUR0000SA0'],"startyear":"2011", "endyear":"2014"})
p = requests.post('https://api.bls.gov/publicAPI/v2/timeseries/data/', data=data, headers=headers)
data = json.loads(p.text)
data
```

```python
# Create pickle/bls and open it for writing, dump bls data to outfile and then close the outfile
```

```python
BLS_PATH = os.path.join("pickle", "bls")
outfile = open(BLS_PATH, 'wb')
pickle.dump(data, outfile)
outfile.close()
```

```python
# Unpickle
```

```python
infile = open('bls', 'rb')
new_data = pickle.load(infile)
infile.close()
```
